```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, roc_auc_score
from sklearn.ensemble import RandomForestClassifier  
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('creditcard1.csv')
df.head(10)

```

```python
df.drop(columns=['Time'], inplace=True)
```

```python
duplicate_values = df.duplicated().sum()
print(f'Number of duplicate rows: {duplicate_values}')

df = df.drop_duplicates()
print(f'Number of rows after dropping duplicates: {df.shape[0]}')
```

```python
classes=df['Class'].value_counts()
print(f'normal ={classes[0]}')
print(f'fraud ={classes[1]}')

### There is a class imbalance... So we have to address this issue
```

```python
title=['normal','fraud']
value=[classes[0],classes[1]]
plt.figure(figsize=(5, 5))
bars = plt.bar(title, value, color=['green','red'])
for bar in bars:
    yval = bar.get_height()  
    plt.text(bar.get_x() + bar.get_width() / 2, yval + 20,  
             str(int(yval)),  
             ha='center', va='bottom', fontweight='bold') 
plt.title('Class Distribution')
plt.xlabel('Transaction Type')
plt.ylabel('Count')
plt.show()
```

```python
df.hist(bins=30, figsize=(15,15))
plt.show()
```

```python
for i in df.columns[:-1]:  
    tmp = df.copy()  
    
    # Discretize the feature into 4 bins (quartiles)
    tmp[f'{i}_bins'] = pd.qcut(df[i], q=4, duplicates='drop')  

    # Compute mean Class value for each bin
    stats = tmp.groupby(f'{i}_bins')['Class'].mean() ## Greater the bars, more is the chance of fraud 

    # Plot
    stats.plot(kind="barh", title=f"{i} vs Class", color='black')
    plt.show()
```

```python
X = df.iloc[:, :-1].values
y = df.iloc[:, -1].values
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
rf = RandomForestClassifier(n_estimators=30, max_depth=30, max_samples=0.2, bootstrap=True, random_state=42)
rf.fit(X_train, y_train)

y_predict = rf.predict(X_test)
y_pred_prob = rf.predict_proba(X_test)  # Get probability scores

accuracy = accuracy_score(y_test, y_predict)
roc_auc = roc_auc_score(y_test, y_pred_prob[:, 1])
print("Accuracy:", accuracy)
print("ROC AUC Score:", roc_auc)
```

```python
df = pd.read_csv('creditcard1.csv')
df.head(10)
```

```python
df.drop(columns=['Time'], inplace=True)
```

```python
# Add Customer_ID column
df.insert(0, 'Customer_ID', range(1, len(df) + 1))

print(df)
```

```python
X = df.iloc[:, :-1].values 
y = df.iloc[:, -1].values
```

```python
y
```

```python
from sklearn.preprocessing import MinMaxScaler
sc = MinMaxScaler(feature_range = (0,1))
X = sc.fit_transform(X)
```

```python
from minisom import MiniSom
som = MiniSom(x=10, y=10, input_len= 30, sigma= 1.0, learning_rate = 0.5)
som.random_weights_init(X)
som.train_random(data = X, num_iteration = 100)
```

```python
from pylab import bone, pcolor, colorbar, plot, show
bone()
pcolor(som.distance_map().T)
colorbar()
markers = ['o', 's']
colors = ['r', 'g']
for i, x in enumerate(X):
    w = som.winner(x)
    plot(w[0] + 0.5,
         w[1] + 0.5,
         markers[y[i]],
         markeredgecolor = colors[y[i]],
         markerfacecolor = 'None',
         markersize = 10,
         markeredgewidth = 2)
show()
```

```python
mappings = som.win_map(X)

# Convert lists to numpy arrays
fraud_1 = np.array(mappings.get((1, 1), []))  # use .get() to handle missing keys
fraud_2 = np.array(mappings.get((4, 1), []))

# Only concatenate if both arrays are non-empty
fraud_arrays = []

if fraud_1.size > 0:  # Check if fraud_1 is not empty
    if fraud_1.ndim == 1:
        fraud_1 = fraud_1.reshape(-1, fraud_1.shape[0])  # reshape to 2D
    fraud_arrays.append(fraud_1)

if fraud_2.size > 0:  # Check if fraud_2 is not empty
    if fraud_2.ndim == 1:
        fraud_2 = fraud_2.reshape(-1, fraud_2.shape[0])  # reshape to 2D
    fraud_arrays.append(fraud_2)

# Only concatenate if we have non-empty arrays
if fraud_arrays:
    frauds = np.concatenate(fraud_arrays, axis=0)
    # Inverse transform the concatenated fraud data
    frauds = sc.inverse_transform(frauds)
else:
    print("No fraud data available for the specified mappings.")    
```
